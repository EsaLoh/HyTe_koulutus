```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
dat = pd.read_csv("https://raw.github.com/EsaLoh/HyTe_koulutus/master/ilmanaikaleimaa.csv")

dat[0:15] #tulostetaan otsikko ja ensimm채inen rivi tarkistuksen vuoksi
```

```
dat2 = dat.rename(columns={'Tunnen itseni onnelliseksi': 'K1', 
                      'Nukuin riittavasti viime yona': 'K2',
                      'Koen usein rauhan tunnetta': 'K3',
                      'Tunnen useamman kerran paivassa iloa, tyytyvaisyytta tai merkityksellisyytta': 'K4',
                      'Minulla on aikaa laheisilleni, ystavilleni ja muille tutuilleni': 'K5',
                      'Minulle on tarkeaa viettaa kahvi- ja lounastauko tyokavereiden seurassa': 'K6',
                      'Tunnen hallitsevani omaa elamaani': 'K7',
                      'Tiesin aamulla, mita tulen tekemaan paivan aikana': 'K8',
                      'Oloni on huva': 'K9',
                      'Soin eilen terveellista ruokaa': 'K10'})

dat2[0:15] #tulostetaan otsikko ja ensimm채inen rivi tarkistuksen vuoksi
```

```
sns.set(style="whitegrid", rc={'figure.figsize':(16,8)})

ax = sns.barplot(x=dat.Nimimerkki, y=dat2.K1, data=dat2, )
```

```
sns.set(style="whitegrid", rc={'figure.figsize':(16,8)})

ax = sns.barplot(x=dat.Nimimerkki, y=dat2.K2, data=dat2)
```

```
sns.set(style="whitegrid", rc={'figure.figsize':(24,8)})

fig, ax =plt.subplots(1,2)
sns.lineplot(x=dat.Nimimerkki, y=dat2.K1, data=dat2, ax = ax[0])
sns.lineplot(x= dat.index, y=dat2.K1, data=dat2, ax = ax[1])

fig.show()

```

```
dat = dat2[['K1', 'K2', 'K3', 'K4', 'K5', 'K6', 'K7', 'K8', 'K9', 'K10']]

corr = dat.corr()
mask = np.triu(np.ones_like(corr, dtype=np.bool))
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
